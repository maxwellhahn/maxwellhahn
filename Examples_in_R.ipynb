```r

# Define the function
f <- function(x) {
  x^2 + 2*x + 4
}

# Define the interval
x_values <- seq(from = -2, to = 20, length.out = 100)

# Calculate y values
y_values <- f(x_values)

# Plot the function
plot(x_values, y_values, type = "l",
     xlab = "x", ylab = "f(x)",
     main = "Plot of f(x) = x^2 + 2x + 4.")
```
